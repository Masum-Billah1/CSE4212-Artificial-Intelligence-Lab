

```
Implementing a simple deep neural network (DNN) for solving the polynomial y = 5x^3 - 8x^2 - 7x + 1
```



#Import necessary module

````python
from tensorflow.keras.layers import Input, Dense, Activation
from tensorflow.keras.models import Model
import numpy as np
import matplotlib.pyplot as plt
````

#a) generate DNN

````python
input = Input((1,), name='Input')
x = Dense(32, activation='relu')(input)
x = Dense(64, activation='relu')(x)
x = Dense(128, activation = 'relu')(x)
output = Dense(1, name='Output_Layer', activation='linear')(x)
dnn_model = Model(input, output, name="DNN")
dnn_model.summary()
````

#b) Generating training samples: range  -20 to +20, normalize: -1 to +1.

````python
x = np.arange(-20,20.01,0.001)
y = (5*(x**3)) - (8*(x**2)) - (7*x) + 1
plt.hist(x)
plt.xlabel('X')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
````

````python
plt.hist(y)
plt.xlabel('Y')
plt.ylabel('Frequency')
plt.title('Histogram-> y')
plt.show()
````

#Normalize X and Y

````python
x_maximum = x.max()
x_minimum = x.min()
x = 2*((x-x_minimum)/ (x_maximum-x_minimum)) - 1
y_maximum = y.max()
y_minimum = y.min()
x = 2*((y-y_minimum)/ (y_maximum-y_minimum)) - 1
````
