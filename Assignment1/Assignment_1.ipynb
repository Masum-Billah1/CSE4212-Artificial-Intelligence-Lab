#Import necessary modules

```python
import numpy as np
import matplotlib.pyplot as plt
```

#Print Step size

```python
x = np.arange(-10, 10.1,0.1)
print(x)
```

#Plot y = wx + b

```python
w = 2
b = 5
y = []
for xi in x:
    y_value = w * xi + b
    y.append(y_value)

plt.xlim([-10, 10])
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of Y = wx + b')
plt.grid('on')
```

#plot y = x^2

```python
y = []
for xi in x:
    y_value = xi**2
    y.append(y_value)

plt.xlim([-10, 10])
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of x^2')
plt.grid('on')
```

#Plot y = 1/ (1 + e^(-x))

```python
y = []
for xi in x:
    y_value = 1/(1+np.exp(xi))
    y.append(y_value)

plt.xlim([-10, 10])
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of 1/ (1 + e^(-x))')
plt.grid('on')
```

#plot y = (e^x - e^(-x)) / (e^x - e^(-x))

```python
y = []
for xi in x:
    y_value = (np.exp(xi) - np.exp(-xi))/(np.exp(xi) - np.exp(-xi))
    y.append(y_value)

plt.xlim([-10, 10])
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of (e^x - e^(-x)) / (e^x - e^(-x))')
plt.grid('on')
```

#plot y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))

```python
w = 2
b = 5
u = []

for xi in x:
    u_value = w * xi + b
    u.append(u_value)

y = []
for ui in u:
  y_value = 1/(1+np.exp(-ui))
  y.append(y_value)

plt.xlim([-10, 10])
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of y = g(f(x))')
plt.grid('on')
```

#Plot y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))

```python
w = 2
b = 5
u = []

for xi in x:
    u_value = w * xi + b
    u.append(u_value)

y = []
for ui in u:
  y_value = (np.exp(ui) - np.exp(-ui))/(np.exp(ui) + np.exp(-ui))
  y.append(y_value)

plt.xlim([-10, 10])
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of y = g(f(x))')
plt.grid('on')
```
