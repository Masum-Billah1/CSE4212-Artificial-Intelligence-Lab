```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = list(range(-10, 11))
print(x)
```

```python
a = 2
b = 5
y = []
for xi in x:
    y_value = a * xi + b
    y.append(y_value)

#plot y = a(xi)+ b
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of Y = ax + b')
```

```python
a = 2
b = 5
x = np.arange(-10, 10.01, 0.01)
y = []
for xi in x:
    y_value = a * np.sin(xi) + b
    y.append(y_value)

#plot y = a(xi)+ b
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of Y = asin(x) + b')
```
